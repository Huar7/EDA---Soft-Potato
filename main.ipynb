```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt


sbn.set_theme(rc={'figure.figsize': (11.7,8.27)})
```

```python
data = pd.read_csv("ds_salaries.csv")
data
```

```python
data.shape
```

```python
data["job_title"].unique()
```

```python
posisi_pekerjaan = data.groupby("job_title").sum()
posisi_pekerjaan = posisi_pekerjaan["sam"]

nama_pekerjaan = data["job_title"].sort_values(ascending=True)

fgr, ax = plt.subplots()

ax.pie(posisi_pekerjaan, labels=nama_pekerjaan.unique(), autopct='%1.1f%%')
plt.show()
```

```python
rata_rata_penghasilan = sum(data["salary_in_usd"]) / len(data["num"])
data_penghasilan = data.groupby("job_title").sum()

print(f"rata rata penghasilan adalah (dalam USD): {rata_rata_penghasilan}")
```

Penghasilan berdasarkan job title nya

```python
data_penghasilan = data_penghasilan["salary_in_usd"]
data_penghasilan
```

Rata - Rata Penghasilannya

```python
rata_rata_data_penghasilan = data_penghasilan.div(len(data_penghasilan))
```

```python

plot1 = sbn.scatterplot(
    rata_rata_data_penghasilan,
)

plot1.set_xticklabels(plot1.get_xticklabels(), rotation=90)
```
